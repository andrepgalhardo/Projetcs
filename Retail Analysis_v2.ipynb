## K-Means with Type and Sales

```python
df_ml = df.copy()
```

```python
df_ml.info()
```

```python
df_ml['Type'].unique().tolist()
```

```python
df_ml['Type'] = df_ml['Type'].str.replace('A', '0')
df_ml['Type'] = df_ml['Type'].str.replace('B', '1')
df_ml['Type'] = df_ml['Type'].str.replace('C', '2')
```

```python
kmeans = KMeans(n_clusters=3).fit(df_ml[['Weekly_Sales', 'Type', 'Size']])
```

```python
three_cluster_df = df_ml[['Weekly_Sales', 'Type', 'Size']].copy(deep=True)
three_cluster_df['Cluster'] = kmeans.labels_
```

```python
plt.scatter(
three_cluster_df.loc[three_cluster_df['Cluster'] == 0]['Size'],
three_cluster_df.loc[three_cluster_df['Cluster'] == 0]['Weekly_Sales'],
c='blue'
)
plt.scatter(
three_cluster_df.loc[three_cluster_df['Cluster'] == 1]['Size'],
three_cluster_df.loc[three_cluster_df['Cluster'] == 1]['Weekly_Sales'],
c='red'
)
plt.scatter(
three_cluster_df.loc[three_cluster_df['Cluster'] == 2]['Size'],
three_cluster_df.loc[three_cluster_df['Cluster'] == 2]['Weekly_Sales'],
c='orange'
)
plt.title('TotalSales vs. OrderCount Clusters')
plt.xlabel('Size')
plt.ylabel('Weekly Sales')
plt.grid()
plt.show()
```
